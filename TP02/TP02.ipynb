Exercice 1 : Exploration de env FrozenLake

```python
import numpy as np
import gym
import plotly.graph_objects as go
```

```python
env = gym.make("FrozenLake-v1", is_slippery=True)
print("Action Space:", env.action_space)
print("Observation Space:", env.observation_space)
```

Implémenintion de Q-Table et l'initialisation

```python
num_states = env.observation_space.n
num_actions = env.action_space.n
Q_table = np.zeros((num_states, num_actions))

# Hyperparameters
alpha = 0.1  # Learning rate
gamma = 0.99  # Discount factor
epsilon = 1.0  # Exploration-exploitation trade-off
epsilon_decay = 0.995
epsilon_min = 0.01
episodes = 5000
```

Q-Learning avec Mise à Jour

```python
rewards_per_episode = []

def train_q_learning():
    global epsilon
    for episode in range(episodes):
        state = env.reset()[0]  
        done = False
        total_reward = 0
        
        while not done:
            if np.random.uniform(0, 1) < epsilon:
                action = env.action_space.sample()  
            else:
                action = np.argmax(Q_table[state, :]) 
            
            new_state, reward, done, _, _ = env.step(action)
            
            Q_table[state, action] = (1 - alpha) * Q_table[state, action] + \
                                     alpha * (reward + gamma * np.max(Q_table[new_state, :]))
            
            total_reward += reward
            state = new_state
        
        epsilon = max(epsilon * epsilon_decay, epsilon_min)
        rewards_per_episode.append(total_reward)
    
    print("Training complete.")

train_q_learning()
```

```python
def plot_rewards():
    fig = go.Figure()
    fig.add_trace(go.Scatter(y=rewards_per_episode, mode='lines', name='Rewards per episode'))
    fig.update_layout(title='Q-Learning Training Progress',
                      xaxis_title='Episodes',
                      yaxis_title='Rewards',
                      template='plotly_dark',
                      width=1000, height=500)
    fig.show()

plot_rewards()

```

```python
def display_q_table():
    print("Final Q-Table:")
    print(Q_table)

display_q_table()
```
